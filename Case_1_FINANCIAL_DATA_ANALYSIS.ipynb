```python
#Gathering Data
#Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load file/table from github.com
df = pd.read_csv("https://raw.githubusercontent.com/agusrusmawan/DS_Bootcamp/main/Facebook_Historical_Stock_Price.csv")
```

```python
#Assessing Data
#displays the first 5 rows of the DataFrame
df.head()
```

```python
#looking for more detailed information about DataFrame
df.info()
```

```python
#From the results above, it can be seen that the amount of data is complete (no missing values) and the data types are all normal
#Checking data duplication
print("Number of duplications: ", df.duplicated().sum())
```

```python
#From the results above, there is no data duplication in the dataframe
#check the summary descriptive statistics from DataFrame
df.describe()
```

```python
##From the above results, it shows that there is no oddity in the summary statistical parameters
```

```python
# Plot closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], color='blue')
plt.title('Historical Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change()

# Calculate 50-day and 200-day moving averages
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()

# Calculate volatility (standard deviation of daily returns)
volatility = df['Daily_Return'].std()

# Print volatility
print('Volatility (standard deviation of daily returns):', volatility)
```

```python
# Plot moving averages and closing prices
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='blue')
plt.plot(df['Date'], df['MA_50'], label='50-Day Moving Average', color='red')
plt.plot(df['Date'], df['MA_200'], label='200-Day Moving Average', color='green')
plt.title('Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
